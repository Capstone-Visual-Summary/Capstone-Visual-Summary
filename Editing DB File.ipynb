```python
import geopandas as gpd # type: ignore
import pandas as pd
from pathlib import Path
from tqdm import tqdm
```

```python

data_path = Path('U:/staff-umbrella/imagesummary/data/Delft_NL/panoids/panoids.geojson')
# data_path = Path('panoids.geojson')
# panoids = gpd.read_file(data_path / 'panoids/panoids.geojson')
panoids = gpd.read_file(data_path)
```

```python
# new_db_unsplit = panoids.copy[['year', 'month', 'dist', 'angle', 'path',]]
# new_db = pd.DataFrame(columns=['img_id','year', 'month', 'path', 'geometry', 'dist', 'angle'])

new_db = pd.DataFrame()

for index, row in tqdm(panoids.iterrows(), total=len(panoids)):
	temp_data_a = {'img_id': index, 'img_id_com': index * 4 + 0, 'part_of_img': 'a', 'year': row['year'], 'month': row['month'],
				'path': row['im_side_a'], 'geometry': row['geometry'], 'dist': row['dist'], 'angle': row['angle']}
	temp_data_front = {'img_id': index, 'img_id_com': index * 4 + 1, 'part_of_img': 'front', 'year': row['year'], 'month': row['month'],
					'path': row['im_front'], 'geometry': row['geometry'], 'dist': row['dist'], 'angle': row['angle']}
	temp_data_b = {'img_id': index, 'img_id_com': index * 4 + 2, 'part_of_img': 'b', 'year': row['year'], 'month': row['month'],
				'path': row['im_side_b'], 'geometry': row['geometry'], 'dist': row['dist'], 'angle': row['angle']}
	temp_data_back = {'img_id': index, 'img_id_com': index * 4 + 3, 'part_of_img': 'back', 'year': row['year'], 'month': row['month'],
				   'path': row['im_back'], 'geometry': row['geometry'], 'dist': row['dist'], 'angle': row['angle']}
	
	new_db = pd.concat([new_db, pd.DataFrame([temp_data_a, temp_data_front, temp_data_b, temp_data_back])], ignore_index=True)

new_geo_db = gpd.GeoDataFrame(new_db, geometry='geometry', crs='EPSG:4326')
```

```python
new_geo_db.to_file('image_info.geojson', driver='GeoJSON')
```

```python
data_path_hoods = Path('cells.geojson')

cells = gpd.read_file(data_path_hoods)
```

```python
new_cells_db = pd.DataFrame()

for index, row in tqdm(cells.iterrows(), total=len(cells)):
	temp_data = {'neighbourhood_id': index, 'geometry': row['geometry'], 'area': row['area']}
	
	new_cells_db = pd.concat([new_cells_db, pd.DataFrame([temp_data])], ignore_index=True)

new_cells_geo_db = gpd.GeoDataFrame(new_cells_db, geometry='geometry', crs='EPSG:4326')
```

```python
display(new_cells_geo_db)
```

```python
new_cells_geo_db.to_file('neighbourhood_info.geojson', driver='GeoJSON')
```
